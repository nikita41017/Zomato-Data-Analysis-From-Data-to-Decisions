```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Zomato data .csv")
```

```python
df.head()
```

```python
df
```

Lets convert the data type of the rate to float and remove denominator 

```python
df.head()
```

```python
def handleRate(value):
    value=str(value).split('/')
    value=value[0]
    return float(value)

df['rate']=df['rate'].apply(handleRate)
print(df.head())
```

```python
df.info()
```

CONCLUSION = there is a no-null value in the data

type of resturant

```python
sns.countplot(x=df['listed_in(type)'])
plt.xlabel("Type of resturant")
```

CONCLUSION- The majority of the resturant fall into the dinning category



```python
grouped_data =df.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes':grouped_data})
plt.plot(result, c ="green", marker="o")
plt.xlabel("TYPE of resturants", c="black", size=20)
plt.ylabel("Votes", c="red", size=20)
```

The majority of resturants received ratings

```python
plt.hist(df['rate'],bins=5)
plt.title("Rating Distribution")
plt.show()
```

CONCLUSION- The majority of resturants received rating ranging from 3.5 to 4.

The majority of couples prefer resturanants with an approximate cost of 300 rupees

```python
couple_data= df['approx_cost(for two people)']
sns.countplot(x=couple_data)
```

wheather online orders receive higher rating than offline orders

```python
plt.figure(figsize=(6,6))
sns.boxplot(x='online_order' , y ='rate', data =df)           
```

```python
pivot_table = df.pivot_table(index='listed_in(type)',columns='online_order', aggfunc='size', fill_value=0)
sns.heatmap(pivot_table, annot=True, cmap="YlGnBu",fmt='d')
plt.title("Heatmap")
plt.xlabel("online order")
plt.ylabel("listed In( Type)")
plt.show()
```

CONCLUSION : DINING resturants primarily accept offline orders, where cafes primarily receive 
online orders. this suggest that clients prefer to place in person orders in person at resturants,
but prefer online orderind at cafes
